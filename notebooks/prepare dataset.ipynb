<h2 align=center>پیش پردازش</h2>
<p align=right>
    .تعدادی داده پرت در داده ما وجود دارد
    <br>
    می توان آنها را به سادگی از دیتافریم حذف کرد.
</p>

```python
df.drop(index=df[(df.Label == 'Side') & (df.Shoulder_to_Hip_Ratio > 1.3)].index, inplace=True)
df.drop(df[df.Eye_to_Ear_Ratio == 0].index, inplace=True)
df
```

```python
df.drop(index=34, inplace=True)
df.drop(index=52, inplace=True)
df
```

<p align=right>
    .با حذف کردن داده ها، اندیس آنها نیز از دیتافریم حذف می شود، پس اندیس ها را دوباره تنظیم می کنیم

```python
df.reset_index(inplace=True)
df
```

<p align=right>
    .متدی که در سلول قبل اجرا شد، یک ستون جدید با نام اندیس به دیتافریم اضافه می کند، پس این ستون اضافی را حذف می کنیم

```python
ax.scatter(df[df.Label == 'Front']['Shoulder_to_Hip_Ratio'], df[df.Label == 'Front']['Eye_to_Ear_Ratio'], marker='x', label='Front Train Data')

ax.scatter(df[df.Label == 'Side']['Shoulder_to_Hip_Ratio'], df[df.Label == 'Side']['Eye_to_Ear_Ratio'], marker='1', label='Side Train Data');
```

```python
ax.scatter(df_test[df_test.Label == 'Front']['Shoulder_to_Hip_Ratio'], df_test[df_test.Label == 'Front']['Eye_to_Ear_Ratio'], marker='v', label='Front Test Data')

ax.scatter(df_test[df_test.Label == 'Side']['Shoulder_to_Hip_Ratio'], df_test[df_test.Label == 'Side']['Eye_to_Ear_Ratio'], marker='h', label='Side Test Data');
```
